```python
import random

user_selection = int(input("What do you choose? Type 0 for Rock, 1 for Paper, 2 for Scissors.\n"))

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

game_options = [rock, paper, scissors]
computer_choice = random.randint(0,2)
user_wins = [[0,2],[1,0],[2,1]]
computer_wins = [[2,0],[0,1],[1,2]]
game = [user_selection, computer_choice]

if (game == user_wins[0] or
        game == user_wins[1] or
        game == user_wins[2]):
    print(game_options[user_selection])
    print(f"Computer chose {game_options[computer_choice]}")
    print("You win!")
elif (game == computer_wins[0] or
        game == computer_wins[1] or
        game == computer_wins[2]):
    print(game_options[user_selection])
    print(f"Computer chose {game_options[computer_choice]}")
    print("Computer win!")
elif user_selection == computer_choice:
    print(game_options[user_selection])
    print(f"Computer chose {game_options[computer_choice]}")
    print("Tied Result! try again")
else:
    print("No a valid option, try again!")
```
