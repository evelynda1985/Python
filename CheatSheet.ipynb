# Day 2

```python
#Subscripting - index of a caracter
print("Hello"[0])
print("Hello"[-1])

# 123_456 =  for numbers, python ignores_
print(123_456 == 123456)

# Math Operations 6/6 = return a float BUT 6//6 = make it as int with a floor round 1.666 = 1
print(6 / 3)
print(6 // 3)

# PEMDASLR Math operation priority
print("(), ", "**, ", "* OR ", "/, ", "+ OR ", "- ")
print(3 * 3 + 3 / 3 -3) # 7
print(3 * (3 + 3) / 3 -3) # 3

# Rounding in python
bmi = 84 / 1.65 ** 2
print(bmi)
print(int(bmi)) #only remove digits after .
print(round(bmi)) #regular round to int
print(round(bmi, 3))

#Accumulative operations
score = 0
score += 1
print(score)

# f-strings avoid the work to convert everything to string for printing it out
score = 0
height = 1.8
is_winning = True
print(f"your score is {score}, your height is {height}. You are winning is {is_winning}")

# 150 * 0.12 + 150 = 150 * 1.12
print(150 * 0.12 + 150)
print(150 * 1.12)
```

# Day 4

```python
#Random
import random
#random integer
random_integer = random.randint(1,10)
print(random_integer)
#random float between 0 and 1
random_number_0_to_1 = random.random()
print(random_number_0_to_1)
#random float
random_float = random.uniform(1,10)
print(random_float)

#Create modules
#Create the code and import it where you want to use it
#file one
my_favorite_number = 3.1415
#file 2
#import my_module
#print(my_module.my_favorite_number)

#Lists [], respect the order
states = ["Delaware", "Pennsylvania"]
print(states[0])

# update a value
states[1] = "pencilvania"
print(states[1])

#adding a value
states.append("New Jersey")
print(states[-1])

#Adding a list
states.extend(["patolandia", "houselandia"])
print(states[-2])

#random index from a list
print(random.choice(states))

#Nested list
fruits = ["Strawberries", "Nectarines", "Apples", "Grapes", "Peaches", "Cherries", "Pears"]
vegetables = ["Spinach", "Kale", "Tomatoes", "Celery", "Potatoes"]
dirty_dozen = [fruits, vegetables]
print(dirty_dozen)
print(dirty_dozen[1][1])

```

# Day 5

```python
#loops
fruits = ["Apple", "Peach", "Pear"]
for fruit in fruits:
    print(fruit)

#sum and max
student_scores = [150, 142, 185, 120, 171, 184, 149, 24, 59, 68, 199, 78, 65, 89, 86, 55, 91, 64, 89]
print(sum(student_scores))
print(max(student_scores))

# range
for number in range(1,10): #it doesn't include 10 [1-9]
    print(number)

#range with intervals
for number in range(1,11,3): #it doesn't include 10 [1-9]
    print(number)

#range of the len
chars = ["a", "b", "c"]
print(f"len chars = {len(chars)}")
for char in range(len(chars)):
    print(char)
```
